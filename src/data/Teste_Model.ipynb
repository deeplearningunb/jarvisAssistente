```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import LinearSVC, LinearSVR
from sklearn.linear_model import LogisticRegression, LinearRegression, SGDClassifier
import numpy as np


from tensorflow.keras.models import Sequential, save_model
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('commands.csv')
```

```python
df = df.sort_index()
```

```python
# x_train, x_test, y_train, y_test = train_test_split(df['command'], df['label'], test_size=0.2, random_state=42)
```

```python
vectorize = CountVectorizer()
x = vectorize.fit_transform(df['command']).toarray()

encoder = LabelEncoder()
y = encoder.fit_transform(df['label'].unique())
y_tras = encoder.transform(df['label'])
encoder_2 = OneHotEncoder(sparse=False)
y_2 = y.reshape(y.shape[0], 1)
y_final = encoder_2.fit(y_2)
y_final_2 = encoder_2.transform(y_tras.reshape(y_tras.shape[0], 1))
y_final_2
```

```python
x.shape
```

```python
y_final_2.shape
```

```python
input_dim = x.shape[1]
model = Sequential()
model.add(Dense(10, activation='relu', input_dim=input_dim))
model.add(Dense(12, activation='sigmoid'))
model.compile(optimizer='adam', loss="binary_crossentropy")

hist = model.fit(x, y_final_2, batch_size=2, epochs=200)
```

```python
test = ['abrir programa', 'passar música', 'tocar música', 'tocar spotify']
x_test = vectorize.transform(test)
```

```python
preds = model.predict(x_test)
```

```python
encoder.inverse_transform([np.argmax(each)for each in preds])
```
